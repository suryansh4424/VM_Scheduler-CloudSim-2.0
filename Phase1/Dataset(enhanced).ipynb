```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Dataset.csv")

# Convert string representation of lists to actual arrays
df['cpu_usage_distribution'] = df['cpu_usage_distribution'].apply(
    lambda x: np.fromstring(x.strip("[]"), sep=' ') if isinstance(x, str) and "[" in x else np.nan
)

# Extract the first value from the array
df['cpu_usage_first'] = df['cpu_usage_distribution'].apply(
    lambda x: x[0] if isinstance(x, np.ndarray) and len(x) > 0 else np.nan
)

# Save cleaned dataset
df.to_csv("cleaned_cpu_usage.csv", index=False)

print("âœ… Dataset cleaned and saved as 'cleaned_cpu_usage.csv' ðŸš€")
```

```python
import pandas as pd
import numpy as np
import ast  # To safely parse dictionary strings

# Load dataset
df = pd.read_csv("Dataset.csv")

# Function to extract only CPU values from dictionary-like strings
def extract_cpu_value(column):
    return df[column].apply(lambda x: ast.literal_eval(x).get("cpus", np.nan) if isinstance(x, str) else np.nan)

# Apply function to extract CPU values
df["resource_request"] = extract_cpu_value("resource_request")
df["average_usage"] = extract_cpu_value("average_usage")
df["maximum_usage"] = extract_cpu_value("maximum_usage")

# Drop columns containing memory usage
df.drop(columns=["cpu_usage_distribution", "tail_cpu_usage_distribution"], inplace=True, errors="ignore")

# Save cleaned dataset
df.to_csv("cleaned_cpu_usage.csv", index=False)

print("âœ… Dataset cleaned and saved as 'cleaned_cpu_usage.csv' ðŸš€")
```

```python
import pandas as pd
import numpy as np
import ast  # To safely parse dictionary strings

# Load dataset
df = pd.read_csv("Dataset.csv")

# Function to extract CPU value and multiply by 1000
def extract_cpu_value(column):
    return df[column].apply(lambda x: ast.literal_eval(x).get("cpus", np.nan) * 1000 if isinstance(x, str) else np.nan)

# Apply function to extract CPU values and scale
df["resource_request"] = extract_cpu_value("resource_request")
df["average_usage"] = extract_cpu_value("average_usage")
df["maximum_usage"] = extract_cpu_value("maximum_usage")

# Convert cpu_usage_distribution and tail_cpu_usage_distribution to numpy arrays
df["cpu_usage_distribution"] = df["cpu_usage_distribution"].apply(
    lambda x: np.fromstring(x.strip("[]"), sep=" ") if isinstance(x, str) else x
)
df["tail_cpu_usage_distribution"] = df["tail_cpu_usage_distribution"].apply(
    lambda x: np.fromstring(x.strip("[]"), sep=" ") if isinstance(x, str) else x
)

# Save cleaned dataset
df.to_csv("Dataset1.csv", index=False)

print("âœ… Dataset cleaned, CPU values extracted & scaled by 1000, distributions preserved, saved as 'cleaned_cpu_usage.csv' ðŸš€")
```
